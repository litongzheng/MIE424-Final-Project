```python
import tensorflow as tf
# import tensorflow_addons as tfa

from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalize pixel values to be between 0 and 1 and center crop to 28x28
train_images, test_images = train_images[:,2:-2, 2:-2, :] / 255.0, test_images[:,2:-2, 2:-2, :] / 255.0

# Standardize mean to 0 and std to 1
train_images, test_images = tf.image.per_image_standardization(train_images), tf.image.per_image_standardization(test_images)

print(tf.math.reduce_mean(train_images[0]), tf.math.reduce_std(train_images[0]))
```

```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    # The CIFAR labels happen to be arrays, 
    # which is why you need the extra index
    plt.xlabel(class_names[train_labels[i][0]])
plt.show()
```
