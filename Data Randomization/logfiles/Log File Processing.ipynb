```python
def extract(d):
  epoch = d.keys()
  train_acc = []
  val_acc = []
  train_loss = []
  val_loss = []
  for i in d.values():
    temp = [j for j in i.values()]
    train_acc.append(temp[0])
    val_acc.append(temp[1])
    train_loss.append(temp[2])
    val_loss.append(temp[3])
  return epoch, train_acc, val_acc, train_loss, val_loss
```
